```python
import yfinance as yf 
import pandas as pd 
import numpy as np 
import matplotlib
```

```python
ticker = "AAPL"
```

```python
# Pour prendre les données de PACKAGE Yfinance on utilise 
# une FONCTION qui est def dans ce package pour EXTRAIRE la donnée 
data = yf.download(ticker , start="2020-01-01", auto_adjust=False)
```

```python
data
```

```python
data.head() # fct pour voir les 5 premières
data.tail() # les 5 dernières 
```

```python
# 2. Calculer les rendements journaliers (Daily Returns)
# On utilise le "Adj Close" (Prix ajusté aux dividendes/splits)
data ['Return'] = data ['Adj Close'].pct_change()

# 3. Calculer la volatilité et le rendement moyen (Annualisés)
# Il y a environ 252 jours de trading dans une année
mean_return_daily = data ['Return'].mean()
volatility_daily = data ['Return'].std()
annualized_return = mean_return_daily * 252
annualized_volatility = volatility_daily * np.sqrt(252)

```

```python
# 4. Calcul du Ratio de Sharpe (supposons taux sans risque = 0)
sharpe_ratio = annualized_return / annualized_volatility 
print(f"--- Analyse pour {ticker} ---")
print(f"Rendement annuel moyen: {annualized_return:.2%}")
print(f"Volatilité annuelle: {annualized_volatility:.2%}")
print(f"Ratio de Sharpe: {sharpe_ratio:.2f}")
```
